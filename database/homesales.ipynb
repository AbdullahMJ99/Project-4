```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Step 1: Read CSV data into a pandas DataFrame
Data_df = pd.read_csv("train.csv")

# Step 2: Create an SQLite database engine
engine = create_engine("sqlite:///database.db")  # Creates an SQLite database named "mydatabase.db"

# Step 3: Insert data into the SQLite database
Data_df.to_sql("mytable", con=engine, index=False, if_exists="replace")
# Replace "mytable" with the desired table name

# Optional: Query the database to verify the insertion
query = "SELECT * FROM mytable LIMIT 5"
result_df = pd.read_sql(query, engine)
print(result_df)
```

```python


# Step 4: Read test.csv into a pandas DataFrame
test_df = pd.read_csv("test.csv")

# Step 5: Insert test data into the SQLite database (append to existing table)
test_df.to_sql("mytable", con=engine, index=False, if_exists="append")

# Optional: Query the database to verify the insertion of test data
query_test = "SELECT * FROM mytable LIMIT 5"
result_test_df = pd.read_sql(query_test, engine)
print("\nTest Data:")
print(result_test_df)
```
